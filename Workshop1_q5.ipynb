```python
def sense(x):
    return x

def simulate(Δt, x, u):
    x += Δt * u
    return x

def control(t, y, kp, ki, xd, x0, ut):
    u = kp * (math.exp(-kp * t)) * (xd - x0) + ki * t * (xd - y)
    return u + ut
```
