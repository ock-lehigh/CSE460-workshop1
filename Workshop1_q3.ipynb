```python
def sense(x):
    return x

def simulate(Δt, x, u):
    x += Δt * u
    return x

def control(t, y):
    ### WRITE YOUR CONTROL POLICY HERE:
    ux = -sin(t)
    uy = cos(t)
    uz = 0.1
    return array([ux, uy, uz])
```

```python
tf = 100.
Δt = 0.1
time = linspace(0., tf, int(tf / Δt) + 1)

x = array([1., 0., 0.])
x_log = [copy(x)]

for t in time:
    y = sense(x)
    u = control(t, y)
    x = simulate(Δt, x, u)
    x_log.append(copy(x))

x_log = array(x_log)

##grid()
##ax.plot3D(x_log[:,0], x_log[:,1], x_log[:,2])


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

plt.plot(x_log[:,0], x_log[:,1], x_log[:,2])
```
