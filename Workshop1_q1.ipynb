```python
def sense(x):
    return x

def simulate(Δt, x, u):
    x += Δt * u
    return x

def control(t, y):
    ### WRITE YOUR CONTROL POLICY HERE:
    ux = - 2 * sqrt(3) * sin(t) - cos(t)
    uy = - 2 * sin(t) + sqrt(3) * cos(t)
    return array([ux, uy])
```

```python
tf = 10.
Δt = 0.1
time = linspace(0., tf, int(tf / Δt) + 1)

x = array([2. * sqrt(3) + 3., 4.])
x_log = [copy(x)]

for t in time:
    y = sense(x)
    u = control(t, y)
    x = simulate(Δt, x, u)
    x_log.append(copy(x))

x_log = array(x_log)

grid()
plot(x_log[:,0], x_log[:,1])
```

```python
import matplotlib.pyplot as plt
from matplotlib import animation
from JSAnimation import IPython_display    
from IPython.display import HTML
    


fig, ax = plt.subplots()

def animate(t):
    ax.clear()
    
    # Path
    plot(x_log[:,0], x_log[:,1], 'r--')
    
    # Initial conditions
    plot(x_log[t,0], x_log[t,1], 'bo')
    
    

anim = animation.FuncAnimation(fig, animate, frames=len(time), interval=60)

HTML(anim.to_jshtml())
```
